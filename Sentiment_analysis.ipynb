```python
import pandas as pd
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk
nltk.download('vader_lexicon')
nltk.download('punkt')
```

```python
sent = SentimentIntensityAnalyzer()
df = pd.read_csv('/content/drive/MyDrive/movie_review.csv', usecols=['sent_id', 'text', 'tag'])
df = df.head(100)
df.head()
```

```python
# Function to extract positive and negative words
def extract_pos_neg_words(text):
    scores = sent.polarity_scores(text)
    positive_words = [word for word in nltk.word_tokenize(text) if sent.polarity_scores(word)['compound'] > 0]
    negative_words = [word for word in nltk.word_tokenize(text) if sent.polarity_scores(word)['compound'] < 0]
    return positive_words, negative_words

# Apply the function to each row in the DataFrame
df[['positive_words', 'negative_words']] = df['text'].apply(lambda x: pd.Series(extract_pos_neg_words(x)))

# Display the DataFrame
df.head()
```

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import pandas as pd

nltk.download('vader_lexicon')
nltk.download('punkt')
# Initialize SentimentIntensityAnalyzer
sent_analyzer = SentimentIntensityAnalyzer()
```

```python
#user defined text
def analyze_sentiment(text):
    # Perform sentiment analysis
    sentiment_scores = sent_analyzer.polarity_scores(text)
    compound_score = sentiment_scores['compound']

    # Extract positive and negative words
    positive_words = [word for word in nltk.word_tokenize(text) if sent_analyzer.polarity_scores(word)['compound'] > 0]
    negative_words = [word for word in nltk.word_tokenize(text) if sent_analyzer.polarity_scores(word)['compound'] < 0]

    # Display the results
    print(f"Sentiment Scores: {sentiment_scores}")
    print(f"Positive Words: {positive_words}")
    print(f"Negative Words: {negative_words}")
    print(f"Overall Sentiment : {compound_score}")

# Example usage:
input_text = "Today i am feeling very sad and depressed."
analyze_sentiment(input_text)
```
